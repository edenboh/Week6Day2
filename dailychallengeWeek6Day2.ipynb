```python
import matplotlib.pyplot as plt

x = df['Happiness Score']
y = df['Economy (GDP per Capita)']

plt.scatter(x,y)
plt.title("Relationship between ‘Economy’ and ‘Happiness Score’.")
plt.xlabel("Happiness Score")
plt.ylabel("Economy")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# First subplot
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('First Subplot')

# Second subplot
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Second Subplot')

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

region_stats = df.groupby("Country")[["Economy (GDP per Capita)", "Health (Life Expectancy)"]].mean()
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(
    region_stats.index,
    region_stats["Economy (GDP per Capita)"],
    alpha=0.7
)
ax1.set_xlabel("Country")
ax1.set_ylabel("Economy (GDP per Capita)")
ax1.tick_params(axis='x', rotation=45)
ax2 = ax1.twinx()

ax2.plot(
    region_stats.index,
    region_stats["Health (Life Expectancy)"],
    marker='o'
)
ax2.set_ylabel("Health (Life Expectancy)")
plt.title("Economy (GDP per Capita) vs Health (Life Expectancy) by Country")
plt.tight_layout()
plt.show()
```
