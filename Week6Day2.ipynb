**Why is data visualization important in data analysis?**

Data visualization is important because it helps people understand data more easily and quickly.

Instead of reading long tables of numbers, visualizations allow us to:

See patterns and trends in the data

Identify outliers or unusual values

Compare different categories or variables

Communicate insights clearly to others

In short, data visualization turns complex data into clear and meaningful information.


**What is the purpose of a line graph in data visualization?**

A line graph is used to show changes or trends over time.

It helps to:

Visualize how a value increases or decreases

Track continuous data such as temperature, sales, or stock prices

Easily compare trends across time periods

Line graphs are especially useful when the order of data matters, such as dates or time intervals.

```python
#exo2
import matplotlib.pyplot as plt
x=['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
y=[72, 74, 76, 80, 82, 78, 75]
plt.plot(x,y)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature Over Week')
plt.show()
```

```python
#exo4
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('sales_data.csv')
sum=df['quantity'].sum()
print("Total quantity sold:", sum)

df.set_index(['category', 'product'], inplace=True)
revenue_by_category = df.groupby('category')['revenue'].sum()
top_category=revenue_by_category.idxmax()
top_revenue = revenue_by_category.max()

print("Top category:", top_category)
print("Revenue generated:", top_revenue)
average_revenue = df["revenue"].mean()
print("Average revenue per sale:", average_revenue)

df.loc[df['date'].between ('2022-01-01', '2022-03-31'), 'quart'] = 'Q1'
df.loc[df['date'].between ('2022-04-01', '2022-06-30'), 'quart'] = 'Q2'
df.loc[df['date'].between ('2022-07-01', '2022-09-30'), 'quart'] = 'Q3'
df.loc[df['date'].between ('2022-10-01', '2022-12-31'), 'quart'] = 'Q4'
revenue_by_quarter = df.groupby('quart')['revenue'].sum()
x=revenue_by_quarter.index
y=revenue_by_quarter.values
plt.bar(x,y)
plt.show()
```

```python
#exo5
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y= [i**2 for i in x]
plt.figure()
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Puissance de 2')
plt.show()

product=['A','B','C','D']
sales=[5, 30, 45,20]
plt.bar(product,sales)
plt.title('product and sales')

colors = ['gold', 'lightblue', 'orange', 'purple']
fruit = ['Apples', 'Bananas', 'Cherries', 'Dates']
like = [40, 30, 20, 10]
plt.figure(figsize=(8, 8))
plt.pie(like, labels=fruit, autopct='%1.1f%%' ,colors=colors)
plt.title('favorite fruits')
plt.show()
```
